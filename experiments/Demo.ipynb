```python
n, _ = data.testSet.input.shape
sample_inds = np.random.choice(np.arange(0, n), 5)
sample = data.testSet.input[sample_inds]

mlpClassifier = MultilayerPerceptron(data.trainingSet, data.validationSet, data.testSet, learningRate=0.25, epochs=10, record_performance=False)
```

```python
for idx in range(len(sample)):
    plt.figure()
    plt.imshow(sample[idx].reshape(28,28))
    
    mlpClassifier._feed_forward(data.testSet.input[sample_inds[idx]])
    label = np.argmax(mlpClassifier.outp)
    print("prediction: " + str(label))
    
    plt.show()
```

```python
mlpClassifier.train(verbose=True)
```
